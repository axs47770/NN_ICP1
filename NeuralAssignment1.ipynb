```python

characters = list(input().strip())


char1 = input().strip()
char2 = input().strip()

if char1 in characters:
    characters.remove(char1)
if char2 in characters:
    characters.remove(char2)

characters.reverse()


result = ''.join(characters)
print(result)
```

```python
print("Enter  total score of student: ")
total = int(input())

if total>=90 and total<=100:
    print("Your Grade is A")
elif total>=80 and total<89:
    print("Your Grade is B")
elif total>=70 and total<79:
    print("Your Grade is C")
elif total>=60 and total<69:
    print("Your Grade is D")
elif total>=0 and total<60:
    print("Your Grade is F")
else:
    print("Invalid Input!")
```
